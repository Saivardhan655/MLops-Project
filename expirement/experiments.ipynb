```python
def assign_label(hour): 
    if hour >= 0 and hour < 6: 
        return 4
    elif hour >= 6 and hour < 12: 
        return 1 
    elif hour >= 12 and hour < 18: 
        return 2 
    elif hour >= 18 and hour <= 24: 
        return 3
```
